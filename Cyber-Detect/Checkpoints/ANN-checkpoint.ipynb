```python
import pandas as pd
import numpy as np
#Visualization
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv('preprocessed_dataset.csv')
```

```python
dataset
```

```python
dataset.fillna(dataset.mean(), inplace=True)
```

```python
mean_value=dataset['duration'].mean()
```

```python
dataset['duration'] = dataset['duration'].replace(0.0, np.nan)
```

```python
dataset['duration'] = dataset['duration'].fillna(mean_value)
```

```python
dataset
```

```python
dataset = dataset.reset_index(drop=True)
dataset = dataset.drop('attack_category', axis=1)
```

```python
dataset = dataset.fillna(dataset.mean())
dataset
```

```python
if dataset['hot'].isnull().any():
    print("There are missing values in column 'column_name'.")
else:
    print("There are no missing values in column 'column_name'.")
```

```python
null_counts = dataset.select_dtypes('float64').isnull().sum()
print(null_counts)
```

```python
scaler = StandardScaler()
scale = scaler.fit_transform(dataset)
X = pd.DataFrame(scale, columns=dataset.columns)
relevant_features = X[['occurance', 'duration']]
```

```python
dataset.plot(kind='scatter',x='occurance',y='duration') # scatter plot
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(relevant_features)
```

```python
# Determine the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(relevant_features)
    labels = kmeans.labels_
    # Add the cluster labels to the DataFrame
    dataset['cluster_label'] = labels
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Visualize the resulting clusters using a scatter plot
plt.scatter(relevant_features['occurance'], relevant_features['duration'], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('occurance')
plt.ylabel('duration')
plt.show()
```
